# Собранные данные типа: язык и ссылка на статью, где он встретился либо в Abstract, либо в Title.
* Стоит заметить, что для языков, название которые меньше 4-х символов, производился поиск по шаблону "<название языка> language"

```python
df_language_url = pd.read_csv('../data\lang_url_dataset_0709.csv')
print(df_language_url.shape)
df_language_url.head()
```

```python
df_language_url['no_brackets'] = df_language_url.language.apply(lambda x: x.split('(')[0].strip())
df_unique = df_language_url[['no_brackets', 'url']].drop_duplicates(subset=['no_brackets', 'url'])
print(df_unique.shape)
df_unique.head()
```

* Информация о том сколько статей содержат упоминание языка:

```python
df_language_url.language.value_counts()
```

```python
df_statistics = pd.DataFrame(
    {
        'Name': df_language_url.language.value_counts().index.to_list(),
        'Num_of_articles': df_language_url.language.value_counts().to_list()
    }
    )
lang_statistic_df = df_wals.merge(df_statistics, on='Name', how='left')
lang_statistic_df.head()
```
