```python
import json
  
# Opening JSON file
f = open('../data/lower_zeroshot.json')
  
# returns JSON object as 
# a dictionary
data_after_training = json.load(f)
```

```python
good_lr_langs = [i for i, j in data_after_training.items() if len(list(j.keys())) > 1 and i in lr_langs]
bad_lr_langs = [i for i, j in data_after_training.items() if len(list(j.keys())) == 1 and i in lr_langs]
```

```python
len(good_lr_langs)
```

```python
for k, v in data_after_training.items():
    if k in lr_langs and len(v) > 1:
        sorted_v = {k: v for k, v in sorted(v.items(), key=lambda item: item[1])}
        print(k, "|", sorted_v["zero_shot"], "|", list(sorted_v.items())[0], )
```

```python
good_hr_langs = {}
for k, v in data_after_training.items():
    if k in lr_langs:
        for vv in v.keys():
            if vv != "zero_shot" and vv in hr_langs:
                if vv not in good_hr_langs:
                    good_hr_langs[vv] = 0
                good_hr_langs[vv] += 1
sorted_dict = {k: v for k, v in sorted(good_hr_langs.items(), key=lambda item: item[1], reverse=True)}
print(sorted_dict)

best_hr_lang = [i for i, j in sorted_dict.items() if sorted_dict[i] > 0]
```

```python
best_hr_lang
```

```python
df_interesting = minimazed_df[minimazed_df.index.isin(best_hr_lang)]
df_interesting = df_interesting[good_lr_langs]

df_interesting.shape
```

```python
delta_df_interesting = df_interesting.copy()
for c in delta_df_interesting.columns:
    delta_df_interesting[c] = delta_df_interesting[c] - [data_after_training[c]["zero_shot"]]
```

```python
delta_df_interesting
```

```python
fig, ax = plt.subplots(1,1, figsize=(10,20))


sns.heatmap(delta_df_interesting, ax=ax, square=True, cbar=True, cmap="CMRmap_r", vmax=100, vmin=-50, linewidths=1.01, 
            cbar_kws={'shrink': 0.4, 'label': '$\Delta$ perplexity', "pad": 0.005})
ax.set_ylabel('HR languages', fontsize=20, rotation=90)
ax.set_xlabel('LR languages', fontsize=20)
ax.tick_params(axis='y', labelsize=15, rotation=0)
ax.tick_params(axis='x', labelsize=15, rotation=90)
cbar = ax.collections[0].colorbar
# here set the labelsize by 20
cbar.ax.tick_params(labelsize=15)
cbar.ax.tick_params(labelsize=15, axis="y", rotation=0)
ax.figure.axes[-1].yaxis.label.set_size(15)


fig.savefig("../data/heatmap_best.pdf", bbox_inches="tight", dpi=500)
```

```python
from scipy.stats import pearsonr
import numpy as np

correlation_coefficients = []
p_values = []


df1 = delta_df_perplexity.fillna(150)
df2 = token_ratio.copy()

for c in df1.columns:
    c, p = pearsonr(df1[c], df2[c])
    correlation_coefficients.append(c)
    p_values.append(p)

print(np.mean(correlation_coefficients), np.mean(p_values))
```
